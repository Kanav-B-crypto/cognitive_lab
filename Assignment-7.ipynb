```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

roll_number = 102317077
np.random.seed(roll_number)

sales_data = np.random.randint(1000,5001,size=(12,4))

columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

df = pd.DataFrame(sales_data,columns=columns,index=months)
print(df)

print(df.head())
print(df.describe())

total_sales_per_category = df.sum(axis=0)
print(total_sales_per_category)
total_sales_per_month = df.sum(axis=1)
print(total_sales_per_month)

monthly_diff = df.diff()              
avg_growth_per_category = monthly_diff.mean() 
print("\nAverage Sales Growth Between Consecutive Months (per Category):")
print(avg_growth_per_category)

# Apply 15% discount to Clothing category (odd roll number)
df['Clothing'] = df['Clothing'] * 0.85
df['Clothing'] = df['Clothing'].round(2)
print("\nSales Data after applying 15% discount to Clothing:\n", df)


for column in df.columns:
    plt.plot(df.index, df[column], marker='o', label=column)
plt.title("Monthly Sales Trends for Each Category")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


sns.boxplot(data=df)
plt.title("Sales Distribution per Category")
plt.xlabel("Product Category")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#2
import numpy as np
array = np.array([
            [1,-2,3],
            [-4,5,-6]
])

abs_array = np.abs(array)
print(abs_array)
print()

flattened = array.flatten()
percentiles_flat = np.percentile(flattened,[25,50,75])
print(percentiles_flat)
print()

percentiles_col_25 = np.percentile(array, 25, axis=0)
percentiles_col_50 = np.percentile(array, 50, axis=0)  # median
percentiles_col_75 = np.percentile(array, 75, axis=0)
print("25th percentile per column:", percentiles_col_25)
print("50th percentile per column:", percentiles_col_50)
print("75th percentile per column:", percentiles_col_75)

percentiles_row_25 = np.percentile(array, 25, axis=1)
percentiles_row_50 = np.percentile(array, 50, axis=1)  # median
percentiles_row_75 = np.percentile(array, 75, axis=1)
print("25th percentile per row:", percentiles_row_25)
print("50th percentile per row:", percentiles_row_50)
print("75th percentile per row:", percentiles_row_75)
print()
# Flattened array
mean_flat = np.mean(flattened)
median_flat = np.median(flattened)
std_flat = np.std(flattened)
print("Flattened array - Mean:", mean_flat, "Median:", median_flat, "Std Dev:", std_flat)
print()
# Each column (axis=0)
mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)
print("Mean per column:", mean_col)
print("Median per column:", median_col)
print("Std Dev per column:", std_col)

# Each row (axis=1)
mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)
print("Mean per row:", mean_row)
print("Median per row:", median_row)
print("Std Dev per row:", std_row)

print('\n',array)
```

```python
#5
my_set = {10,20,30,40,50}

my_list = list(my_set)
i = 1
j = 3

print("Before swap (list):", my_list)

temp = my_list[i]
my_list[i] = my_list[j]
my_list[j] = temp

print("After swap (list):", my_list)

new_set = set(my_list)
print("Converted back to set:", new_set)
```
