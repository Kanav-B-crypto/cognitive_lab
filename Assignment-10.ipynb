```python
# Required Libraries
import re
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Step 0: Paragraph
paragraph = """Technology is evolving at an incredible pace. From smartphones to artificial intelligence, innovations are transforming every aspect of our lives. 
The Internet has made the world more connected than ever. Modern gadgets are not just convenient but also essential. 
Technology has become a powerful tool in education, healthcare, and communication."""

# Step 1: Convert to lowercase and remove punctuation
clean_text = re.sub(r'[^\w\s]', '', paragraph.lower())
print("Cleaned Text:\n", clean_text)

# Step 2: Tokenize into sentences and words
sentences = sent_tokenize(paragraph)
words_nltk = word_tokenize(clean_text)

print("\nSentences:\n", sentences)
print("\nNLTK Word Tokens:\n", words_nltk)

# Step 3: Compare split() and word_tokenize()
words_split = clean_text.split()
print("\nPython split():\n", words_split)

print("\nDifference:")
print("Extra tokens in word_tokenize():", set(words_nltk) - set(words_split))
print("Extra tokens in split():", set(words_split) - set(words_nltk))

# Step 4: Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_nltk if word not in stop_words]
print("\nFiltered Words (No Stopwords):\n", filtered_words)

# Step 5: Frequency Distribution
fdist = FreqDist(filtered_words)
print("\nWord Frequency Distribution (Most Common):")
print(fdist.most_common())

# Plot the frequency distribution
fdist.plot(10, title="Top 10 Word Frequencies (Excluding Stopwords)")
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required NLTK data
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Sample paragraph (same as before)
paragraph = """Technology is evolving at an incredible pace. From smartphones to artificial intelligence, innovations are transforming every aspect of our lives. 
The Internet has made the world more connected than ever. Modern gadgets are not just convenient but also essential. 
Technology has become a powerful tool in education, healthcare, and communication."""

# 1. Extract all words with only alphabets
words_alpha = re.findall(r'\b[a-zA-Z]+\b', paragraph)
print("Words with only alphabets:\n", words_alpha)

# 2. Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_alpha if word.lower() not in stop_words]
print("\nFiltered Words (no stopwords):\n", filtered_words)

# 3. Perform stemming with PorterStemmer
ps = PorterStemmer()
stemmed_words = [ps.stem(word.lower()) for word in filtered_words]
print("\nStemmed Words:\n", stemmed_words)

# 4. Perform lemmatization with WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word.lower()) for word in filtered_words]
print("\nLemmatized Words:\n", lemmatized_words)

# 5. Compare stemmed and lemmatized outputs
print("\nComparison of Stemmed vs Lemmatized:")
for original, stem, lemma in zip(filtered_words, stemmed_words, lemmatized_words):
    print(f"{original:15} -> Stemmed: {stem:10} | Lemmatized: {lemma}")

# Explanation:
explanation = """
Stemming is a crude heuristic process that chops off word endings to reduce words to their base or root form, which might not be a valid word (e.g., 'transforming' -> 'transform').
Lemmatization is a more informed process that uses vocabulary and morphological analysis to return the base or dictionary form of a word (e.g., 'transforming' -> 'transform').

Use stemming when speed and simplicity are priorities, and you don't mind approximate roots.
Use lemmatization when you want more accurate, linguistically correct base forms, especially in tasks requiring proper understanding of the words.
"""
print(explanation)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import numpy as np

# 3 short example texts (e.g., product reviews or news headlines)
texts = [
    "The smartphone has an excellent camera and battery life.",
    "This laptop is fast, lightweight, and has a sleek design.",
    "I love the noise cancellation feature on these headphones."
]

# 1. Bag of Words with CountVectorizer
count_vect = CountVectorizer()
bow_matrix = count_vect.fit_transform(texts)
print("Bag of Words (CountVectorizer) Feature Names:\n", count_vect.get_feature_names_out())
print("\nBag of Words Matrix:\n", bow_matrix.toarray())

# 2. TF-IDF with TfidfVectorizer
tfidf_vect = TfidfVectorizer()
tfidf_matrix = tfidf_vect.fit_transform(texts)
feature_names = tfidf_vect.get_feature_names_out()

print("\nTF-IDF Matrix:\n", np.round(tfidf_matrix.toarray(), 3))

# 3. Print top 3 keywords by TF-IDF for each text
print("\nTop 3 keywords per text by TF-IDF scores:")
for i, row in enumerate(tfidf_matrix.toarray()):
    top_indices = row.argsort()[-3:][::-1]  # indices of top 3 TF-IDF scores
    top_keywords = [(feature_names[idx], row[idx]) for idx in top_indices]
    print(f"\nText {i+1}: \"{texts[i]}\"")
    for word, score in top_keywords:
        print(f"  - {word}: {score:.3f}")
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')

# 1. Two short texts about different technologies
text1 = """Artificial intelligence (AI) enables machines to learn from data and make decisions. 
It is used in speech recognition, autonomous vehicles, and medical diagnosis."""

text2 = """Blockchain technology is a decentralized ledger system that provides security and transparency. 
It is widely used in cryptocurrencies and smart contracts."""

# Preprocessing function: lowercase + remove punctuation + tokenize
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    tokens = word_tokenize(text)
    return tokens

tokens1 = preprocess(text1)
tokens2 = preprocess(text2)

print("Tokens Text 1:", tokens1)
print("Tokens Text 2:", tokens2)

# 2a. Jaccard Similarity (using sets)
set1 = set(tokens1)
set2 = set(tokens2)

jaccard_sim = len(set1.intersection(set2)) / len(set1.union(set2))
print(f"\nJaccard Similarity: {jaccard_sim:.3f}")

# 2b. Cosine Similarity using TF-IDF
texts = [text1, text2]
tfidf_vect = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vect.fit_transform(texts)

cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
print(f"Cosine Similarity: {cosine_sim:.3f}")

# 2c. Analysis
analysis = """
Analysis:
- Jaccard similarity measures overlap of unique words ignoring frequency and semantics. It is simple but may not capture nuanced similarity.
- Cosine similarity uses TF-IDF weighted vectors considering word importance and frequency, providing a richer similarity measure.
- In this case, cosine similarity better reflects the textual context and semantic closeness, especially for technical texts with related terms.
"""

print(analysis)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.probability import FreqDist
from nltk.stem import PorterStemmer,WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
text1 = "Artificial Intelligence enables machines to learn from data and perform tasks that require human intelligence."
text2 = "Blockchain is a decentralized technology that ensures secure and transparent transactions using distributed ledgers."
stop_words=set(stopwords.words('english'))
def preprocess(text):
  return [word for word in word_tokenize(text) if word.isalpha() and word not in stop_words]

tokens1=set(preprocess(text1))
tokens2=set(preprocess(text2))
intersection=tokens1.intersection(tokens2)
union=tokens1.union(tokens2)
jaccard_similiarity=len(intersection)/len(union)
print(f"Jaccard Similarity is {jaccard_similiarity}")

tfidf=TfidfVectorizer()
tfidfmatrix=tfidf.fit_transform([text1,text2])
cos_sim=cosine_similarity(tfidfmatrix[0:1],tfidfmatrix[1:2])[0][0]
print(f"Cosine Similarity is {cos_sim}")
```

```python
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample reviews
reviews = [
    "I absolutely love this phone! The battery life is amazing and the camera quality is superb.",
    "The laptop is okay, but it heats up quickly and the keyboard is not comfortable.",
    "Service was excellent, very friendly staff and fast response time.",
    "I am disappointed with the product. It stopped working after a week.",
    "The headphones are decent but nothing special."
]

# 1 & 2: Analyze sentiment & classify
classified_reviews = []
positive_texts = []

for review in reviews:
    analysis = TextBlob(review)
    polarity = analysis.sentiment.polarity
    subjectivity = analysis.sentiment.subjectivity

    if polarity > 0.1:
        sentiment = "Positive"
        positive_texts.append(review)
    elif polarity < -0.1:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    classified_reviews.append((review, polarity, subjectivity, sentiment))

# Print results
for rev, pol, subj, sent in classified_reviews:
    print(f"Review: {rev}")
    print(f" Polarity: {pol:.2f}, Subjectivity: {subj:.2f}, Sentiment: {sent}\n")

# 3: Create word cloud for positive reviews
if positive_texts:
    text = " ".join(positive_texts)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title("Word Cloud for Positive Reviews")
    plt.show()
else:
    print("No positive reviews to generate word cloud.")
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# 1. Your own paragraph (~100 words)
text = """
Artificial intelligence and machine learning are revolutionizing many industries. 
From healthcare to finance, AI algorithms improve decision-making and automate tasks. 
Deep learning, a subset of machine learning, uses neural networks to model complex patterns. 
Data scientists and engineers continuously develop new techniques to enhance AI capabilities. 
As technology advances, AI's impact will grow, changing how we live and work every day.
"""

# 1. Tokenize text
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1
print(f"Total unique words: {total_words}")

# 2. Create input sequences (n-grams)
input_sequences = []
for line in text.split('\n'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(2, len(token_list)+1):
        n_gram_seq = token_list[:i]
        input_sequences.append(n_gram_seq)

# Pad sequences
max_seq_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')

# Create predictors and label
input_sequences = np.array(input_sequences)
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

# One-hot encode labels
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=total_words)

# 3. Build a simple LSTM model
model = Sequential()
model.add(Embedding(total_words, 10, input_length=max_seq_len-1))
model.add(LSTM(50))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Train the model
model.fit(X, y, epochs=100, verbose=0)

# 4. Generate new text from seed word
def generate_text(seed_text, next_words=20):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_seq_len-1, padding='pre')
        predicted = model.predict(token_list, verbose=0)
        predicted_word_index = np.argmax(predicted, axis=1)[0]
        
        # If predicted word is unknown, break early
        if predicted_word_index == 0:
            break
        
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_word_index:
                output_word = word
                break
        
        seed_text += " " + output_word
        
        # Stop if punctuation or sentence end (optional)
        if output_word in ['.', '!', '?']:
            break
    return seed_text

# Provide your seed word here
seed = "artificial"
generated_text = generate_text(seed, next_words=20)
print("\nGenerated Text:")
print(generated_text)
```
