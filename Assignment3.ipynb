```python
#6
import pandas as pd
sample_dataset={
    'Employee_ID':[101,102,103,104,105],
    'Name':['Alice','Bob','Charlie','Diana','Edward'],
    'Department':['HR','IT','IT','Marketing','Sales'],
    'Age':[29,34,41,28,38],
    'Salary':[50000,70000,65000,55000,60000],
    'Years of Experience':[4,8,10,3,12],
    'Joining Date':['2020-03-15','2017-07-19','2013-06-01','2021-02-10','2010-11-25'],
    'Gender':['Female','Male','Male','Female','Male'],
    'Bonus':[5000,7000,6000,4500,5000],
    'Rating':[4.5,4.0,3.8,4.7,3.5],
}
df=pd.DataFrame(sample_dataset)
print(df)
#a
print("Shape of the data frame:",df.shape)#number of rows and columns
#b
print("Summary of the DataFrame that includes the data types and non-null counts for each column:",df.info())
#c
print("Descriptive statistics:",df.describe())
#d
print("Display the first 5 rows",df.head())
print("Display the last 3 rows",df.tail(3))
#e
print("Average salary of employees",df['Salary'].mean())
print("The total bonus paid to all employees",df['Bonus'].sum())
print("The youngest employee's age",df['Age'].min())
print("The highest performance rating",df['Rating'].max())
#f
sorted_df=df.sort_values(by='Salary',ascending=False)
print("The DataFrame by the Salary column in descending order:")
print(sorted_df)
#g
df.loc[df['Rating']>=4.5,'Performance_Category']='Excellent'
df.loc[(df['Rating']>=4) & (df['Rating']<4.5),'Performance_Category']='Good'
df.loc[df['Rating']<4,'Performance_Category']='Average'
print(df)
#h
missing_values=df.isnull()
missing_count=df.isnull().sum()
print("Missing values in each column: ")
print(missing_count)
#i
df.rename(columns={'Employee_ID':'ID'},inplace=True)
print(df)
#j
print("All employees who have more than 5 years of experience and Belong to the IT department")
ans_df=df.loc[(df['Years of Experience']>5) & (df['Department']=='IT')]
print(ans_df)
#k
df['Tax']=0.1*df['Salary']
print("Modified")
print(df)
#l
modified_csv_path = "modified_employees.csv"
df.to_csv(modified_csv_path, index=False)

print(f"Modified DataFrame saved to {modified_csv_path}")
```
