```python
#Q1
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Part I: Randomized Sales Data Generation
roll_number = 102317077  
np.random.seed(roll_number)
categories = ['Electronics','Clothing','Home & Kitchen','Sports']
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales_data = np.random.randint(1000,5000,size=(12,4))

df = pd.DataFrame(sales_data,columns=categories,index=months)

# Part II: Data Manipulation and Analysis
print(df.head())
print(df.describe())
total_sales_category = df.sum()
print("Total Sales per Category:")
print(total_sales_category)
df['Total Sales'] = df.sum(axis=1)
print("Total Sales per Month:")
print(df['Total Sales'])
df['Growth Rate'] = df['Total Sales'].pct_change()*100

discount_category = 'Electronics' if roll_number % 2 == 0 else 'Clothing'
df[discount_category] *= 0.9 if discount_category == 'Electronics' else 0.85

# Part III: Visualizations

plt.figure(figsize=(10, 6))
for category in categories:
    plt.plot(df.index, df[category], marker='o', label=category)
plt.title("Monthly Sales Trends")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x=categories, y=total_sales_category.values, palette='husl')
plt.title("Total Sales per Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
#Q2
import numpy as np
array = np.array([[1, -2, 3], [-4, 5, -6]])

absolute_values = np.abs(array)
print("Element-wise Absolute Values:\n", absolute_values)
flattened_array = array.flatten()
print("\nflattened array",flattened_array)

percentiles_flattened = np.percentile(flattened_array, [25, 50, 75])
percentiles_columns = np.percentile(array, [25, 50, 75], axis=0)
percentiles_rows = np.percentile(array, [25, 50, 75], axis=1)

print("\n25th, 50th, 75th Percentiles (Flattened):", percentiles_flattened)
print("\n25th, 50th, 75th Percentiles (Columns):\n", percentiles_columns)
print("\n25th, 50th, 75th Percentiles (Rows):\n", percentiles_rows)


mean_flattened = np.mean(flattened_array)
median_flattened = np.median(flattened_array)
std_flattened = np.std(flattened_array)

mean_columns = np.mean(array,axis=0)
median_columns = np.median(array,axis=0)
std_columns = np.std(array,axis=0)

mean_rows = np.mean(array,axis=1)
median_rows = np.median(array,axis=1)
std_rows = np.std(array,axis=1)

print("\nMean, Median, Std Dev (Flattened):", (mean_flattened,median_flattened,std_flattened))
print("\nMean (Columns):",mean_columns)
print("Median (Columns):",median_columns)
print("Std Dev (Columns):",std_columns)
print("\nMean (Rows):",mean_rows)
print("Median (Rows):",median_rows)
print("Std Dev (Rows):",std_rows)
```

```python
#Q4
def swap_elements(lt,pos1,pos2):
    temp = lt[pos1]
    lt[pos1] = lt[pos2]
    lt[pos2] = temp
    return lt

my_list = [10,20,30,40,50]
pos1,pos2 = 1,3 

print("Original List:",my_list)
swapped_list = swap_elements(my_list,pos1,pos2)
print("List after swapping:",swapped_list)
```

```python
#Q5
def swap_elements_in_set(s,elem1,elem2):
    # Convert set to list
    lt = list(s)
    
    # Check if both elements exist in the set
    if elem1 in lt and elem2 in lt:
        index1,index2 = lt.index(elem1),lt.index(elem2)
        lt[index1],lt[index2] = lt[index2],lt[index1]
        s = set(lt)
        return s
    else:
        return "Elements not found in the set!"

my_set = {10, 20, 30, 40, 50}
print("Original Set:",my_set)
swapped_set = swap_elements_in_set(my_set,20,50)
print("Set after swapping:",swapped_set)
```
