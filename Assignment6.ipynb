```python
#Q1
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'r--', label=f'y = {M} * x^2')  # Red dashed line
plt.plot(x, y2, 'b-', label=f'y = {M} * sin(x)')  # Blue solid line

plt.legend()
plt.grid(True)
plt.title("Graphs of y = M * x^2 and y = M * sin(x)")
plt.xlabel("x values")
plt.ylabel("y values")

plt.show()
```

```python
#Q2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']
scores = [85, 90, 78, 88, 92]

df = pd.DataFrame({'Subject': subjects, 'Score': scores})
plt.figure(figsize=(8, 6))
colors = sns.color_palette("husl", len(subjects))  # Different colors for each bar
ax = sns.barplot(x='Subject', y='Score', data=df, palette=colors)
for index, value in enumerate(scores):
    ax.text(index, value + 1, str(value), ha='center', fontsize=12, fontweight='bold')

plt.title("Subject Scores", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Q3
import numpy as np
import matplotlib.pyplot as plt

roll_number = 102317077  
np.random.seed(roll_number)
data = np.random.randn(50)

cumulative_sum = np.cumsum(data)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(cumulative_sum, 'b-', marker='o', label='Cumulative Sum')
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)
axes[0, 0].legend()

axes[0, 1].scatter(range(50), data, c='r', marker='x', label='Random Noise')
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value")
axes[0, 1].grid(True)
axes[0, 1].legend()

axes[1, 0].axis('off')
axes[1, 1].axis('off')
plt.tight_layout()
plt.show()
```
