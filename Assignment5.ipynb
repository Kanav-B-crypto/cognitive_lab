```python
#Q1
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
total_sum = np.sum(gfg)
row_wise_sum = np.sum(gfg, axis=1)
column_wise_sum = np.sum(gfg, axis=0)
print("Sum of all elements: ",total_sum)
print("Sum of all elements row-wise: ",row_wise_sum)
print("Sum of all elements column-wise: ",column_wise_sum)
```

```python
#Q2
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
sorted_indices = np.argsort(array)
smallest_elements = np.sort(array)[:4]
largest_elements = np.sort(array)[-5:]
print("Sorted array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 smallest elements:", smallest_elements)
print("5 largest elements:", largest_elements)


array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer_elements = array[array == array.astype(int)]
float_elements = array[array != array.astype(int)]
print("Integer elements only:", integer_elements)
print("Float elements only:", float_elements)
```

```python
#Q3
import numpy as np
#initials: KB
X = ord('K') + ord('B')  # 75 + 66 = 141
sales = np.array([X, X+50, X+100, X+150, X+200])
print("X (ASCII sum of initials):", X)
print("Sales dataset:", sales)

tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * tax_rate
print("Personalized tax rate:", tax_rate)
print("Tax applied to each sales value:", taxed_sales)

discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Original sales:", sales)
print("Discounted sales:", discounted_sales)

weekly_sales = np.tile(sales, (3, 1))  
week_multipliers = np.array([1.00, 1.02, 1.04]).reshape(3, 1)  
adjusted_sales = weekly_sales * week_multipliers
print("Original weekly sales:\n", weekly_sales)
print("\nSales after 2% increase per week:\n", adjusted_sales)
```

```python
#Q4
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y1 = x**2                
y2 = np.sin(x)           
y3 = np.exp(x)           
y4 = np.log(np.abs(x) + 1)  
plt.figure(figsize=(10, 6))

plt.plot(x, y1, label=r'$Y = x^2$', linestyle='-', color='b')
plt.plot(x, y2, label=r'$Y = \sin(x)$', linestyle='--', color='g')
plt.plot(x, y3, label=r'$Y = e^x$', linestyle='-.', color='r')
plt.plot(x, y4, label=r'$Y = \log(|x| + 1)$', linestyle=':', color='m')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Mathematical Functions Plot")
plt.legend()
plt.grid(True)
plt.show()
```
